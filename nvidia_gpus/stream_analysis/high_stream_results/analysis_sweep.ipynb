```python
#########
import glob
import csv
#Going through CSV file, finding index of average count, adding all counts of functions together
combined_data = {'inst_exe':[],'ipc_count':[],'array_size':[], 'warps_launched':[]}

for filen in glob.glob("stream_ipc*"):
    #print (filen)
    array_size = filen.split('_')[2].split('N')[1]
    block_size = filen.split('_')[3].split('B')[1]
    
    # Use the array size and block size to open files for all other metrics with the same array-block setting
    
    file = open(filen ,'r')
    combined_data['array_size'].append(array_size)
    title = {}
    fdata = csv.reader(file)
    startevent = False
    startipc = False
    events = 0
    ipc = 0.0
    cnt = 0.0
    for line_split in fdata:
        
        #print(line_split)
        lsplt = (len(line_split) > 0)
        
        if startevent:
            if lsplt:
                avg_index = title['Avg']
                events += int(line_split[ avg_index ])
            else: startevent = False

        if startipc:
            if lsplt:
                avg_index = title['Avg']
                ipc += float(line_split[ avg_index ])
                cnt += 1.0
            else: startipc = False

        elif lsplt and line_split[0] == 'Device' and line_split[1] == 'Kernel':
            if line_split[3] == 'Event Name':
                startevent = True
            else:
                startipc = True
            idx = 0
            for term in line_split:
                title[term] = idx
                idx += 1
            
            
    #print(combined_data['ipc_count'])  
    combined_data['warps_launched'].append(events)
    val = ipc / max (1.0, cnt)
    combined_data['ipc_count'].append(val) # Average of values
    file.close()

print (combined_data)
```

```python
# Getting all values of array and block combos. available
combined_data = {'inst_exe':[],'ipc_count':[],'array_size':[], 'warps_launched':[], 'block_size':[], 'local_mem':[]}

for filen in glob.glob("stream_*"):
    #print (filen)
    array_size = filen.split('_')[2].split('N')[1]
    block_size = filen.split('_')[3].split('B')[1].split('.')[0]
    combined_data['array_size'].append(array_size)
    combined_data['block_size'].append(block_size)
print("All combos of array and block sizes only: \n")
print (combined_data)
```

```python
types = {'inst_':'inst_exe','instwarp':'warps_launched','ipc':'ipc_count','local':'local_mem'}
place = 0
for filen in glob.glob("stream_*"):
    #print (filen)
    if place < len(combined_data['array_size']):
        if combined_data['array_size'][place] in filen and combined_data['block_size'][place] in filen:
            place += 1
            file = open(filen ,'r')
            title = {}
            fdata = csv.reader(file)
            startevent = False
            startipc = False
            events = 0
            inst_exe = 0.0
            cnt = 0.0
            for line_split in fdata:
                #print(line_split)
                lsplt = (len(line_split) > 0)

                if startevent:
                    if lsplt:
                        avg_index = title['Avg']
                        
                        events += int(line_split[ avg_index ])
                    else: startevent = False

                if startipc:
                    if lsplt:
                        avg_index = title['Avg']
                        #print (avg_index)
                        if '%'in line_split[ avg_index ]:
                            inst_exe += float(line_split[ avg_index ].split('%')[0])
                        else:
                            inst_exe += float(line_split[ avg_index ])
                        cnt += 1.0
                    else: startipc = False

                elif lsplt and line_split[0] == 'Device' and line_split[1] == 'Kernel':
                    if line_split[3] == 'Event Name':
                        startevent = True
                    else:
                        startipc = True
                    idx = 0
                    for term in line_split:
                        title[term] = idx
                        idx += 1


                #print(combined_data['ipc_count'])  
                #combined_data['warps_launched'].append(events)
                val = inst_exe / max (1.0, cnt)
                for k,v in types.items():
                    if k in filen:
                        combined_data[v].append(val) # Average of values
file.close()
print (combined_data)
```
